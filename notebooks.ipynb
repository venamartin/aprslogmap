```python
import folium
import csv
import datetime
import geopy.distance
```

```python
# load the log file
track = []
digipeaters = []
first = None
with open('aprs.log', encoding="utf8") as csvfile:
    aprs = csv.reader(csvfile)
    for row in aprs:
        timestamp = row[1]
        callsign = row[3]
        digipeater = row[4]
        lat = row[10]
        lon = row[11]
        #print(timestamp, callsign, digipeater, lat, lon)
        if "KN6NCG" in callsign:
            if first is None:
                first = (float(lat), float(lon))
                track.append((float(lat), float(lon)))
                digipeaters.append({'digipeater':digipeater, 'pos':(float(lat), float(lon))})
            elif geopy.distance.geodesic(first, (float(lat), float(lon))).km < 100:
                track.append((float(lat), float(lon)))
                digipeaters.append({'digipeater':digipeater, 'pos':(float(lat), float(lon))})
                
print (track)
        
print(datetime.datetime.fromtimestamp(1713369706))
```

```python
# plot the track on the map
m = folium.Map(location=track[0], zoom_start=11)

folium.PolyLine(track, tooltip="Coast").add_to(m)

my_colors = {"K6LY-3":'blue', "FMTPK":'red'}
for d in digipeaters:
    #print(d['digipeater'])
    
    folium.Marker(
        location=d['pos'],
        #tooltip="Click me!",
        popup=d['digipeater'],
        icon=folium.Icon(color=my_colors.get(d['digipeater'], 'orange'),icon="dot"),
    ).add_to(m)

m
```
